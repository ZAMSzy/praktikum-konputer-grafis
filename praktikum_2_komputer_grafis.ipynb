```python
"""
CIRCLES
"""
import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75,75,50,-50])

plt.axis('on')
plt.grid(True)

plt.arrow(0,0,20,0,head_lenght=4,head_width=3,color='k')
plt.arrow(0,0,0,20,head_lenght=4,head_width=3,color='k')

plt.text(16,-2,'x')
plt.text(-5,17,'y')

#green circle
xc=0
yc=0
r=40

p1=0*np.pi/180
p2=360*np.pi/180

dp=(p2-p1)/100
xlast=xc+r*np.cos(p1)
ylast=yc+r*np.sin(p1)
for p in np.arange(p1,p2+dp,dp):
      x=xc+r*np.cos(p)
      y=yc+r*np.sin(p)
      if p > 90*np.pi/180 and p < 270*np.pi/180:
            plt.plot([xlast,x],[ylast,y],color='g',lineStyle=':')

      else:
            plt.plot([xlast,x],[ylast,y],color='g')
      xlast=x
      ylast=y

plt.scatter(xc,yc,s=15,color='g')
```

```python
"""
ELLIPSES
"""


import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75,75,50,-50])

plt.axis('on')
plt.grid(True)



plt.arrow(0,0,60,0, head_length=4,head_width=3,color='k')
plt.arrow(0,0,0,45, head_length=4,head_width=3,color='k')

plt.text(58,-3,'x')
plt.text(-5,44,'y')
#red ellips
a=40
b=20.

p1=0
p2=180*np.pi/180
dp=.2*np.pi/180

xplast=a
yplast=0
for p in np.arange(p1,p2,dp):
      xp=np.abs(a*b*(b*b+a*a*(np.tan(p))**2.)**-.5)
      yp=np.abs(a*b*(a*a+b*b/(np.tan(p)**2.))**-.5)
      if p > np.pi/2:
          xp=-xp
      plt.plot([xplast,xp],[yplast,yp],color='r')
      plt.plot([xplast,xp],[-yplast,-yp],color='r')
      xplast=xp
      yplast=yp
#green ellips
a=20.
b=40.
xp1=-a
xp2=a
dx=.1

xplast=-a
yplast=0
for xp in np.arange(xp1,xp2,dx):
      yp=b*(1-xp**2./a**2.)**.5
      plt.plot([xplast,xp],[yplast,yp],linewidth=1,color='g')
      plt.plot([xplast,xp],[-yplast,-yp],linewidth=1,color='g')
      xplast=xp
      yplast=yp
      plt.plot([xplast,a],[yplast,0],linewidth=1,color='g')
      plt.plot([xplast,a],[-yplast,0], linewidth=1,color='g')

#blue ellips
a=5.
b=15.
p1=0
p2=180*np.pi/180
dp=.2*np.pi/180

for p in np.arange(p1,p2, dp):
      xp=np.abs(a*b*(b*b+a*a*(np.tan(p))**2.)**-.5)
      yp=np.abs(a*b*(a*a+b*b/(np.tan(p)**2.))**-.5)
      if p > np.pi/2:
          xp=-xp
      plt.plot([xp,xp],[yp,-yp],linewidth=1,color='b')

plt.show()
```

```python
"""
2DROCTANGEL
"""
import matplotlib.pyplot as plt
import numpy as np

plt.axis([-10,150,100,-10])
plt.axis('on')
plt.grid(True)
#
plt.arrow(0,0,40,0,head_length=4,head_width=2,color='b')
plt.arrow(0,0,0,40,head_length=4,head_width=2,color='k')
plt.text(30,-3,'Xg',color='b')
plt.text(-8,34,'Yg',color='b')
#
xc=75
yc=50
plt.plot([xc-40,xc+60],[yc,yc],linewidth=1,color='grey')
plt.plot([xc,xc],[yc-40,yc+45],linewidth=1,color='grey')
plt.text(127,48,'X')
plt.text(70,90,'Y')
plt.scatter(xc,yc,s=20,color='k')
#
def rotz(xp,yp,rz):
    c11=np.cos(rz)
    c12=-np.sin(rz)
    c21=np.sin(rz)
    c22=np.cos(rz)
    xpp=xp*c11+yp*c12 #--relative to xc,yc
    ypp=xp*c21+yp*c22
    xg=xc+xpp #--relative to xg,g
    yg=yc+ypp
    return [xg,yg]
#
xp1=-20
xp2=+20
xp3=+20
xp4=-20
yp1=-5
yp2=-5
yp3=+5
yp4=+5

plt.text(50,45,'1') # --label
plt.text(97,45,'2')
plt.text(97,57,'3')
plt.text(50,57,'4')
plt.scatter(xp1+xc, yp1+yc, s=10,color='k')
plt.scatter(xp2+xc,yp2+yc,s=10,color='k')
plt.scatter(xp3+xc,yp3+yc,s=10,color='k')
plt.scatter(xp4+xc, yp4+yc,s=10,color='k')
#
xg1=xc+xp1 # -corner coordinates in Xg,Yg system
yg1=yc+yp1
xg2=xc+xp2
yg2=yc+yp2
xg3=xc+xp3
yg3=yc+yp3
xg4=xc+xp4
yg4=yc+yp4
#
xg=[xg1,xg2,xg3,xg4,xg1]
yg=[yg1,yg2,yg3,yg4,yg1]

plt.plot((xg),(yg),color='k')
#
rz=45
rz=rz*np.pi/180
#
xp=xp1
yp=yp1
[xg,yg]=rotz(xp,yp,rz)
[xg1,yg1]=[xg,yg]
#
xp=xp2
yp=yp2
[xg,yg]=rotz(xp,yp,rz)
[xg2,yg2]=[xg,yg]
#
xp=xp3
yp=yp3
[xg,yg]=rotz(xp,yp,rz)
[xg3,yg3]=[xg,yg]
#
xp=xp4
yp=yp4
[xg,yg]=rotz(xp,yp,rz)
[xg4,yg4]=[xg,yg]
#
xg=[xg1,xg2,xg3,xg4,xg1]
yg=[yg1,yg2,yg3,yg4,yg1]

plt.plot(xg,yg,color='r')

plt.show()
```
